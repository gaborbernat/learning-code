# Dictionaries

A dictionary (also referred to as map synonym) is a data structure that allows us to store mappings. It's made up of a key and a value. For example see how one could define the mapping from numeric number to English words representing that number.

```python
number_to_text = {
    2: "two",
    3: "three",
    1: "one",
}
```

Here is a visual representation of these mappings:

```python
objviz(number_to_text)
```

To put a new element into a dictionary you can use the following syntax:

```python
number_to_text[1.5] = 0.1
# first start with writing the name of the dictionary
# then use the square brackets to specify the key where to put it
# finally putting the equals sign allow us to specify the value we want to put into the dictionary right afterwards
```

After the above line runs we can see our dictionary now contains four elements:

```python
objviz(number_to_text)
```

To remove an element you can use the following syntax:

```python
del number_to_text[3]
```

If we look at the dictionary we can now see it no longer contains the deleted key (together with its value associated):

```python
objviz(number_to_text)
```

Note the keys are unique, you can't have two entries having the same key. The mapping defined by the dictionary must be unambiguous, so when someone says give me the value for key ``1``, there must be exactly one value to be returned. If we would allow to have two entries with the same key value this could not be done.  

We can get the value associated with a key by using the following syntax:

```python
number_to_text[2]
# name of the dictionar and then we specify the key within the square brackets
```

# Dictionaries with list values
In the above examples the keys are number, while the values are strings. While generally the key must be a primitive type (think of numbers, text, booleans), the value can be anything. It can even be another dictionary, or list. Let's consider the following example demonstrating this, first with the values being a list.

```python
dog_category_to_breeds = {
    "toy": ["yorkie", "pug", "maltese",],
    "hound": ["beagle",],
}
```

If we visualize this we can see that the value is no longer one level, but two:

```python
objviz(dog_category_to_breeds)
```

Lists are also a kinda a mapping, the difference between them and a dictionary is that while inside a dictionary the key can be any primitive content (string, boolean, etc) for the list the keys are always numbers that start with 0 and  go up one by one all the way to the size of the list. The visualization above hints to that, by putting the index numbers of the list elements above the values within the list.

What we learned above about how to get a value from within a dictionary still applies. We can use the ``__name_of__dictionary__[key]`` format which will return us the value:

```python
dog_category_to_breeds["toy"]
```

Or visualized.

```python
objviz(dog_category_to_breeds["toy"])
```

However now the value no longer is just a string, but it's the list. Accessing elements from a list is done by also using the square bracket notation, with the difference that the keys now are the index of the elements within the list. So if want to get the second dog type from within the toy group we can first get the toy group, and then get the second element:

```python
toy_group_breeds = dog_category_to_breeds["toy"]
toy_group_breeds[1]
```

Or we can do all this in a single line, chaining up the square brackets:

```python
dog_category_to_breeds["toy"][1]
# note list indexes start with 0, so when we want to get n-th element we need to lookup n-1 element
```

# Nested dictionaries


We can even have dictionaries where the value is a dictionary, hence having nested dictionaries. Here's an example of this:

```python
dog_category_to_breed_to_rating = {
    "toy": {"yorkie": 10, "pug": 10, "malteese": 10, "Chinese Crested": 8},
    "hound": {"beagle": 9},
}
```

Here's this visualized:

```python
objviz(dog_category_to_breed_to_rating)
```

As you can see now we have a mapping within a mapping, to get the rating of the beagle, we first need to lookup the
hound group, and then then the beagle itself.

```python
hound_group_breed_to_rating = dog_category_to_breed_to_rating["hound"]
hound_group_breed_to_rating["beagle"]
```

Here's visualized the result of looking up the hound group:

```python
print(hound_group_breed_to_rating)
objviz(hound_group_breed_to_rating)
```

As you can see applying the square bracket to the dictionary results in having only the nested dictionary.

Or in a more condensed one line form where we chain the square bracktes:

```python
dog_category_to_breed_to_rating["hound"]["beagle"]
```

The above technique works with arbitrary deep nested dictionaries, for example given:

```python
capitals = {
    "Europe": {
        "UK": {
            "England": "London",
            "Northern Ireland": "Belfast",
            "Wales": "Cardiff",
            "Scotland": "Edinburgh",
        },
        "Hungary": "Budapest",
    },
    "North America": {"Canada": "Toronto", "USA": "Washington",},
    "Asia": {"Japan": "Tokyo", "Thailand": "Bangkok",},
}
```

```python
objviz(capitals)
```

Now try to get the capital of ``Northern Ireland`` from the dictionary, you can enter your code below and press SHIFT+ENTER to execute.

```python
capitals["Europe"]["UK"]["Northern Ireland"]
```

# Operations on dictionaries

## checking if something's in there



To get an element from a dictionary we can use the the square bracket, but what happens if that element is not there?

```python
capitals["Africa"]
```

As we see will throw an error, so if we're not sure if something is in the dictionary we should first check if it is in there, for this we can use the **in** operator and an **if** statement:

```python
if "Europe" in capitals:
    print(capitals["Europe"])
else:
    print("Africa is not within capitals")
```

Or alternatively use the ``get`` function on the dictionary, this will return the value if the key is within the 
dictionary or otherwise the value ``None``.

```python
print(capitals.get("Africa"))  # example when it's not there
```

```python
print(capitals.get("Asia"))  # example when it's there
```

How can we find out if Japan is within our nested dictionary from above? We can just combine the in operator we learned here together with the dictionary lookup from before (assuming we know that Japan is within Asia):

```python
asia = capitals["Asia"]
if "Japan" in asia:
    print(asia["Japan"])
else:
    print("Japan is not within capitals")
```

## Going through the keys

We can use the **for** loop to go though the values of a dictionary:

```python
objviz(capitals)
```

```python
for contintent in capitals:
    print(contintent)
```

Note we can get a list of the dictionary keys also by using the **keys** method on the dictionary:

```python
print(capitals.keys())
```

This returns them in a list like structure, however in this form we can only use them to traverse (think for loop) through them, not to index it:

```python
# this will work
for continent in capitals.keys():
    print(continent)
```

```python
# this will not:
capitals.keys()[0]
```

Note however that you can always make them a proper list by wrapping it within a ``list()``:

```python
list(capitals.keys())
```

```python
keys_as_list = list(capitals.keys())
print(keys_as_list)
```

```python
keys_as_list[0]
```

## dictionary values

```python
objviz(capitals)
```

How can we find out if our dictionary contains the capital of Japan, but without knowing that Japan is within Asia? 
We would need to go through all continents, and check if Japan is in their dictionary. Note in this case we don't really care about the continent name, we only care about the values of the dictionary. In this case we can get access
directly to the values of the dictionary by using the **values** function on the dictionary:


```python
objviz(capitals.values())
```

As you can see in this case we  no longer have access to the capital, but instead each element of this values
view provides as access directly to the countries, so we can just check if the dictionaries at that position contain japan or not:

```python
search_for = "Japan"
has_japan = False
for countries in capitals.values():
    if search_for in countries:
        has_japan = True
        print("found " + search_for)
if not has_japan:
    print("could not find " + search_for)
```

Would we searched for another country this would have failed:

```python
search_for = "Silky"
has_japan = False
for countries in capitals.values():
    if search_for in countries:
        has_japan = True
        print("found " + search_for)
if not has_japan:
    print("could not find " + search_for)
```

As you can see sometimes the only way to find something is to go through the nested dictionaries and have if statements that check if something is there or not. In such cases having a ``has_{something}`` variable can
be used to store if the search was success or not, even once the search finished.

## Having multiple dictionaries if we want to lookup by multiple things

If we're searching for something multiple times, it might be worth to have multiple lookup tables/dictionaries. Consider the following:

```python
dogs = [
    {"bitch": "yes", "name": "Molly", "type": "Yorkie"},
    {"bitch": "no", "name": "Silky", "type": "Yorkie"},
    {"bitch": "yes", "name": "Magic", "type": "Pug"},
]
```

```python
objviz(dogs)
```

How can we find all the Yorkies? By using some for loop, here we collecting in a new list.

```python
bitches = []
for dog in dogs:
    if dog["bitch"] == "yes":
        bitches.append(dog)
print(bitches)
```

```python
objviz(bitches)
```

Another way we can do this is we built a dictionary of lists that separates bitches and dogs in separate lists:

```python
bitch_status_to_dogs = {}

for dog in dogs:
    bitch_status = dog["bitch"]

    # first if the current bitch status is not yet a dictionary let's add it to it, with an empty list
    if bitch_status not in bitch_status_to_dogs:
        bitch_status_to_dogs[bitch_status] = []

    # get the current status dogs list
    current_status_dogs = bitch_status_to_dogs[bitch_status]
    # add to the current status dogs list the current one
    current_status_dogs.append(dog)

    # now similarly how we chained together square brackets we can also chain lookuop in dictionary and then append
    # the following line does in one line what before we did in two
    # current_status_dogs[bitch_status].append(dog)
```

```python
objviz(bitch_status_to_dogs)
# note here we have a dictionary where the values are list, that contain dog dictionaries
```

Using the same logic we can build a breed to dogs mappings too:

```python
breed_to_dogs = {}

for dog in dogs:
    breed_type = dog["type"]
    if breed_type not in breed_to_dogs:
        breed_to_dogs[breed_type] = []
    breed_to_dogs[breed_type].append(dog)
```

```python
objviz(breed_to_dogs)
```

Now we have two dictionaries, one that allows to quickly lookup by bitch status ``bitch_status_to_dogs`` and one that allows to lookup by breed type aka ``breed_to_dogs``.

Note that even though we have two dictionaries to lookup by, each with their own set of dogs; the dog information itself is still only once present in the computer. If we change the dogs list, that will also automatically modify the information for both of our dictionaries:

```python
objviz(dogs)
```

```python
dogs[0]["name"] = "Meg"
```

```python
objviz(dogs)
```

```python
objviz(breed_to_dogs)
```

```python
objviz(bitch_status_to_dogs)
```

This can also hurts us though, would we change the bitch status, our bitch status dictionary would no longer be correct:

```python
dogs[0]["bitch"] = "no"
```

```python
objviz(bitch_status_to_dogs)
```

Just watch out when you have multiple dictionaries pointing to same structure, as changing elements might mean you need to rebuild/fix some of the dictionaries.
